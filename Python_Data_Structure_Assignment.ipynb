```python
# 1.Discuss string slicing and provide examples?


# Ans:

# In python, strings support indexing and slicing. Every individual units of a string (character) has a specific index value in the string, which helps in accessing required set of characters out of a given string.

# Strings are indexed starting from 0 to (length of the string - 1) in left to right direction and -vely indexed starting from -1 to -(length of the string) in right to left direction.

# String slicing is a technique used to extract a portion of a string by specifying a range of indices.

# The general syntax for string slicing is "string[start:stop:step]"; in which start index, stop index and step value are optional at cases as per requirement.

# In string slicing, the start index value is inclusive and the stop index value is exclusive in nature.

# Default value for start index is 0 and for stop index is length of the string and the default step value is 1.

# Examples of String Slicing:

# Basic slicing:

text = "Python code"
t_slice = text[1:5]
print(t_slice)


text = "Hello, World!"
substring = text[:5]
print(substring)

substring = text[7:]
print(substring)


# Slicing with Negative Indexing:

text = "Python code"
t_slice = text[-6:-1]
print(t_slice)

# Slicing with step value other than 1:

text = "Python code"
t_slice = text[0:7:2]
print(t_slice)

# Reversing a string using slicing:

text = "abcdefg"
r_text = text[::-1]
print(r_text)


# with -ve step value

text = "Python code"
substring = text[11:6:-1]
print(substring)

# Slicing is highly efficient in Python and is widely used in various text manipulation operations.
```

```python
# 3.Describe how to access, modify and delete elements in a list with examples?


# Ans:

#Python lists allow easy access, modification, and deletion of elements as it is a ordered, heterogenous and sequential data structure. Here's how to perform these operations:

# a) Accessing Elements in a List

# we can access list elements using their index. Python lists support both positive indexing (starting from 0) and negative indexing (starting from -1 for the last element).

# Examples of accessing elements from a list:

demo_list = [10, 2+3j, 30, "Python", 70]

# Positive indexing
print(demo_list[0])
print(demo_list[3])

# Negative indexing
print(demo_list[-1])
print(demo_list[-3])

# using slicing
print(demo_list[1:4])
print(demo_list[:3])

# b) Modifying Elements in a List

# Since lists are mutable, we can modify their elements by assigning a new value to a specific index after creation.

# Examples of Modifying elements:

demo_list = [10, 20, 30, 40, 50]

# Modify element at index 2
demo_list[2] = 35
print(demo_list)

# Modify a range of elements (slice assignment)
demo_list[1:3] = [25, 45]
print(demo_list)


# c) Deleting Elements in a List

# There are several ways to delete elements from a list: using del, remove(), and pop().

# "del" is used to remove an element or a range of elements by their index.

Ex: demo_list = [10, 20, 30, 40, 50]

# Delete element at index 1
del demo_list[1]
print(demo_list)  # Output: [10, 30, 40, 50]

# Delete a range of elements
del demo_list[1:3]
print(demo_list)

# remove() removes the first occurrence of a specific value.

demo_list = [10, 20, 30, 20, 40]

# Remove the first occurrence of 20
demo_list.remove(20)
print(demo_list)

# pop() removes and returns the element at a specified index. If no index is provided, it removes the last element.
#python.

demo_list = [10, 20, 30, 40, 50]

# Remove and return the element at index 2
popped_item = demo_list.pop(2)
print(popped_item)
print(demo_list)

# Remove and return the last element
popped_item = demo_list.pop()
print(popped_item)
print(demo_list)


# clear() removes all elements from the list.

demo_list = [10, 20, 30]
demo_list.clear()
print(demo_list)

# These features make Python lists flexible for managing collections of data in various scenarios.
```

```python
# 4.Compare and contrast tuples and lists with examples.



# Ans:

# Both tuples and lists are used to store collections of heterogenous data in Python, but they have significant differences in their functionality, usage, and behavior. Here are some comparisons and differences between tuples and lists.

# a)Mutability

# Lists: Mutable (can be modified after creation).
# Tuples: Immutable (cannot be modified after creation).

# List (mutable)
list = [1, 2, 3]
list[1] = 20  # Modifying an element
print(list)  # Output: [1, 20, 3]

# Tuple (immutable)
tuple1 = (1, 2, 3)
# tuple1[1] = 20  # Raises TypeError: 'tuple' object does not support item assignment

# b)Syntax

# Lists: Defined using square brackets [].
# Tuples: Defined using parentheses ().

sample_list = [10, 20, 30]  # List
sample_tuple = (10, 20, 30)  # Tuple

# c)performance

# Lists: Slower than tuples because they are mutable and require additional memory for dynamic resizing.
# Tuples: Faster than lists, as they are immutable and have a smaller memory footprint.

# d)Use Case

# Lists: Used when you need a collection of items that can change (i.e., when elements are added, removed, or modified).
# Tuples: Used when you need a fixed collection of items (i.e., when data should not change).

# Example:
# List: A collection of grade of employees in an organization (which changes regularly after a certain time period)
# Tuple: A collection of employee ids of the employees of an organization (Always remains constant)

# e)Methods

# Lists: Support many built-in methods like append(), remove(), pop(), and sort().
# Tuples: Only have two methods: count() and index().

# Ex:
list1 = [10, 20, 30]
list1.append(40)
print(list1)  # Output: [10, 20, 30, 40]

list1.pop()
print (list1)

# Tuple methods
tuple1 = (10, 20, 20, 30)
print(tuple1.count(20))  # Output: 2 (counts occurrences of 20)
print(tuple1.index(30))  

# f)Size

# Lists: Consume more memory due to their dynamic nature.
# Tuples: Generally use less memory than lists since they are immutable.

# g) Iteration

# Both lists and tuples can be iterated over using loops, but tuples are typically faster due to their immutability.

# h)Nested Structures

# Lists: Can contain other lists (nested lists).
# Tuples: Can contain other tuples or lists (nested tuples or lists).

# Ex:
list1 = [[1, 2], [3, 4]]
print(list1)

# Nested tuple
tuple1 = ((1, 2), (3, 4))
print(tuple1)

# Tuple containing a list
l_tuple = (1, [2, 3])
l_tuple[1][0] = 10  # we can modify the list inside the tuple
print(l_tuple)  # Output: (1, [10, 3])

# So, comparatively tuple type data structure works faster than list type data structure in python.
```

```python
# 5. Describe key features of sets and provide examples of their use.



# Ans:

# A set is an unordered collection of unique elements. Sets are useful when we want to store distinct items and perform common operations like union, intersection, and difference.

# The elements of set do not have index values and they have no definite order.

# Below are the key features of sets:

# a)Unordered:

# Sets do not maintain any order of elements. When you print a set or iterate over it, the order of elements might differ from the order in which they were added.

# Elements of set don't have any index value. So, we can't access them by their position.

# Ex:
sample_set = {3, 1, 2}
print(sample_set)

# b)Unique Elements

# Sets don't allow duplicate values as it's member. Sets automatically remove duplicate values.

# Ex:
sample_set = {1, 2, 2, 3, 4, 4}
print(sample_set)

# c)Mutable

# Sets are mutable, means we can add or remove elements after the set is created.

# Ex:
sample_set = {1, 2, 3}
sample_set.add(4)  # Adding a new element
print(sample_set)

sample_set.remove(2)  # Removing an element
print(sample_set)

# d)contains Immutable Elements

# Sets can only store immutable (hashable) elements, such as numbers, strings, or tuples. Lists and dictionaries, which are mutable, cannot be elements of a set.

# Ex:
sample_set = {1, 2, (3, 4)}  # Tuples are allowed
print(sample_set)

# my_set = {1, 2, [3, 4]}  # Raises TypeError: unhashable type: 'list'

# e)No Indexing or Slicing

# Since sets are unordered, they do not support indexing or slicing. we cannot access elements by their position.

# Ex:
my_set = {1, 2, 3}
# print(my_set[0])  # Raises TypeError: 'set' object is not subscriptable

# f)Set Operations

# Union: Combines all unique elements from two sets.

#ex:
set1 = {1, 2, 3}
set2 = {3, 4, 5}

union_set = set1 | set2
print(union_set)

# Intersection: Finds elements that are common in both sets.

# ex:
set1 = {1, 2, 3}
set2 = {2, 3, 4}

# intersection_set = set1 & set2
print(intersection_set)

# Difference: Finds elements present in the first set but not in the second.

# ex:
set1 = {1, 2, 3}
set2 = {2, 3, 4}

difference_set = set1 - set2
print(difference_set)

# Symmetric Difference : Finds elements present in either set, but not in both

#ex:
set1 = {1, 2, 3}
set2 = {2, 3, 4}

symmetric_diff_set = set1 ^ set2
print(symmetric_diff_set)

# g)Set Comprehension

# We can create sets using set comprehensions.

# ex:
my_set = {x**2 for x in range(5)}
print(my_set)

# h)Built-in Methods

# Sets have several useful built-in methods for common operations:

# add(): Adds an element to the set.

# ex:
my_set = {1, 2}
my_set.add(3)
print(my_set)

# remove(): Removes a specific element (raises KeyError if the element is not found).

# ex:
my_set = {1, 2, 3}
my_set.remove(2)
print(my_set)

# discard(): Removes an element without raising an error if the element is not found.

# ex:
my_set = {1, 2, 3}
my_set.discard(2)
print(my_set)

# pop(): Removes and returns an arbitrary element (since sets are unordered, any element can be removed).

# ex:
my_set = {1, 2, 3}
popped_element = my_set.pop()
print(popped_element)
print(my_set)  # Remaining set

#clear(): Removes all elements from the set.

# ex:
my_set = {1, 2, 3}
my_set.clear()
print(my_set)

# i)Examples of Set Use Cases

# a)Removing Duplicates : Sets are commonly used to remove duplicates from a list or other iterable.

# ex:
my_list = [1, 2, 2, 3, 3, 4]
my_set = set(my_list)
print(my_set)

# b)Membership Testing : Sets are optimized for checking membership using the in keyword.

# ex:
my_set = {1, 2, 3, 4}
print(3 in my_set)
print(5 in my_set)

# c)Set Operations on Large Data: When working with large datasets, set operations like union, intersection, and difference can be efficiently used to compare groups of data.

# Sets are ideal when dealing with large collections of unique data and when you need efficient membership testing or set operations.

```

```python
# 7. Describe how to add, modify and delete items in a dictionary with examples.



# Ans:

# In Python, dictionaries are mutable collections of key-value pairs, where each key is unique. You can add, modify, and delete items in a dictionary easily.

# A)Adding Items to a Dictionary

# we can add new key-value pairs by simply assigning a value to a key that does not already exist in the dictionary.

#ex:
# Initial dictionary
student = {'name': 'Alice', 'age': 20}

# Add a new key-value pair
student['major'] = 'Computer Science'

print(student)

# we can also use the update() method to add multiple key-value pairs or to update existing ones.

#ex:
# Initial dictionary
student = {'name': 'Alice', 'age': 20}

# Addition of multiple new key-value pairs
student.update({'major': 'Computer Science', 'gpa': 3.8})

print(student)


# B)Modifying Items in a Dictionary

#To modify an item in a dictionary, we assign a new value to an existing key.

#ex:
# Initial dictionary
student = {'name': 'Alice', 'age': 20, 'major': 'Computer Science'}

# Modify the value of the 'age' key
student['age'] = 21

print(student)

# The update() method can also be used to modify multiple items in a dictionary.

#ex:
# Initial dictionary
student = {'name': 'Alice', 'age': 21, 'major': 'Computer Science'}

# Modify multiple key-value pairs
student.update({'age': 22, 'gpa': 3.9})

print(student)

# C)Deleting Items from a Dictionary

# There are several ways to delete items from a dictionary, depending on the requirement.

# we can delete a key-value pair using the del keyword. This will remove both the key and its corresponding value from the dictionary.

# ex:
# Initial dictionary
student = {'name': 'Alice', 'age': 22, 'major': 'Computer Science', 'gpa': 3.9}

# Delete the 'gpa' key-value pair
del student['gpa']

print(student)


# The popitem() method removes and returns the last key-value pair added to the dictionary (in Python 3.7+). This is useful for dealing with the most recent additions.

# ex:
# Initial dictionary
student = {'name': 'Alice', 'age': 22, 'major': 'Computer Science'}

# Remove the last inserted key-value pair
last_item = student.popitem()

print(student)

print(last_item)

#The clear() method removes all items from the dictionary, leaving it empty.

#ex:
# Initial dictionary
student = {'name': 'Alice', 'age': 22}

# Clear all key-value pairs
student.clear()

print(student)


#With these methods, we can easily add, modify, and delete items from dictionaries in Python based on your requirement.
```

```python
# 8.Discuss the importance of dictionary key being immutable and provide examples.



# Ans:

# In Python, dictionary keys must be immutable to ensure the reliability and efficiency of operations like key lookups, and maintaining data integrity.

# This requirement means that dictionary keys must be of data types that cannot be changed after they are created, such as strings, numbers, or tuples (with immutable elements). Lists, sets, and other mutable objects cannot be used as dictionary keys.

# Immutable keys ensure that once a key-value pair is stored, the key’s hash value remains the same, ensuring fast and consistent retrieval.

# Immutability ensures that keys remain unchanged, which helps maintain the integrity of the dictionary. If a key could be changed (as would be the case with mutable objects), it would break the logical relationship between keys and their values, potentially leading to data corruption.

# Allowing mutable keys would create inconsistencies in the dictionary. For instance, if a mutable key is modified after insertion, finding the key again would become impossible because the hash value would no longer match its original location in memory.

# Examples of Immutable Data Types as Keys:

# a)Strings

# Strings are immutable in Python, making them ideal for use as dictionary keys.

# ex:
my_dict = {'name': 'Alice', 'age': 25}
print(my_dict['name'])

# b) Numbers

# Integers and floating-point numbers are also immutable and can be used as dictionary keys.

#ex:
prices = {1: 'apple', 2: 'banana', 3: 'cherry'}
print(prices[2])

# c)Tuples (Immutable Elements)

# Tuples are immutable, meaning their values cannot be changed after creation. They can be used as keys in a dictionary, provided that the tuple itself contains only immutable objects (e.g., strings, numbers).

# ex:
points = {(1, 2): 'point A', (3, 4): 'point B'}
print(points[(1, 2)])

# Mutable objects, such as lists and sets, cannot be used as dictionary keys because their values can change over time. This change would disrupt the hash table, leading to problems with Data retrieval, Data consistency.

# Hence Using immutable objects like strings, numbers, and tuples ensures dictionaries remain performant and reliable, while preventing potential issues with mutable types like lists and sets.
```
